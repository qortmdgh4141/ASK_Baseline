```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.random(1024)
y = np.random.random(1024)
```

```python
array = np.concatenate([x[:,np.newaxis],y[:,np.newaxis]], axis=1)
array.reshape(1024,2)
array.shape
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x,y)

plt.scatter(x[200],y[200], c ='red')

```
